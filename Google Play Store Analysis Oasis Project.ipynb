```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv(r'C:\Users\USER\Downloads\archive (4)\datasets\Google Play Store dataset.csv')
```

```python
# Display the first few rows
df.head()
```

```python
# Check data types and summary statistics
df.info()
df.describe(include='all')
```

```python
# Data Cleaning and Correction
# Remove duplicates and handle missing values
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
```

```python
# Correct data types
df['Reviews'] = df['Reviews'].astype(int)
df['Size'] = df['Size'].replace('Varies with device', None)
```

```python
# Check and handle missing values
df.isnull().sum()
```

```python
df.info()
```

```python
# Category Exploration
category_counts = df['Category'].value_counts()
plt.figure(figsize=(15, 8))
sns.barplot(x=category_counts.index, y=category_counts.values, palette='viridis')
plt.title('App Distribution Across Categories')
plt.xticks(rotation=90)
plt.show()
```

```python
# Metrics Analysis
# Average Rating per Category
avg_ratings = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)
plt.figure(figsize=(15, 8))
sns.barplot(x=avg_ratings.index, y=avg_ratings.values, palette='coolwarm')
plt.title('Average Rating by Category')
plt.xticks(rotation=90)
plt.show()
```

```python
# Popularity Trends
install_counts = df['Installs'].str.replace('[+,]', '', regex=True).astype(int)
df['Installs'] = install_counts
top_installed_apps = df.nlargest(10, 'Installs')

plt.figure(figsize=(15, 8))
sns.barplot(x=top_installed_apps['App'], y=top_installed_apps['Installs'], palette='mako')
plt.title('Top 10 Most Installed Apps')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
price = df.Price != 0
price.sum()
```

```python
unique_price_count = df['Price'].nunique()
print(f"Number of unique Price values: {unique_price_count}")
```

```python
# Pricing Trends
# Group prices into intervals of 100
df['Price_Interval'] = pd.cut(df['Price'], bins=range(0, int(df['Price'].max()) + 100, 100))
price_interval_counts = df['Price_Interval'].value_counts().sort_index()

plt.figure(figsize=(15, 8))
sns.barplot(x=price_interval_counts.index.astype(str), y=price_interval_counts.values, palette='Set3')
plt.title('Pricing Trends Grouped by Intervals of 100')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Price Intervals')
plt.ylabel('Number of Apps')
plt.show()
```
