```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Drop rows with missing 'host_name' or 'last_review' and 'name'
df = df.dropna(subset=['name', 'host_name', 'last_review',])
```

```python
# Step 5. Standardization
# Standardize categorical fields (e.g., consistent capitalization).
# Format date columns.
```

```python
# Standardize text fields
df['name'] = df['name'].str.title()
df['host_name'] = df['host_name'].str.title()
```

```python
# Convert 'last_review' to datetime
df['last_review'] = pd.to_datetime(df['last_review'])
```

```python
# Step 6. Outlier Detection
# Use statistical methods like Z-scores or IQR to identify outliers in numerical fields (e.g., price, minimum_nights).
import numpy as np
```

```python
# Detect outliers in 'price' using IQR
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
```

```python
# Filter out outliers
outlier_condition = (df['price'] < (Q1 - 1.5 * IQR)) | (df['price'] > (Q3 + 1.5 * IQR))
print(f"Number of outliers in 'price': {outlier_condition.sum()}")
```

```python
# Remove outliers
df = df[~outlier_condition]
```
